1. Line Plot with Matplotlib Objective: Create a line plot showing the temperature variation over a
week. Steps: 1. Data Preparation: • Create a list of days of the week (e.g., Monday to Sunday). • Create a
list of corresponding temperatures (e.g., [22, 24, 21, 23, 25, 26, 27]).
2. Plotting: • Use Matplotlib to create a line plot. • Add grid lines, title, labels for the x-axis and y-axis. •
Add a legend to the plot.



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday',
'Saturday', 'Sunday']
temperatures = [22, 24, 21, 23, 25, 26, 27]
```

```python
# Line plot
plt.plot(days, temperatures, label='Temperature Variation',
marker='o')
plt.title('Temperature Variation Over a Week')
plt.xlabel('Day of the Week')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.legend()
plt.show()
```

Practical Assignment 3: Interactive Scatter Plot with Plotly
Objective: Create an interactive scatter plot showing the relationship between two variables.
Steps:
1. Data Preparation:
• Use the Iris dataset available in Seaborn.
2. Plotting:
• Use Plotly to create an interactive scatter plot.
• Add hover effects to display data points.

```python
# Load data
import seaborn as sns
df = sns.load_dataset('iris')
print(df)
```

```python
# Interactive scatter plot
# Interactive scatter plot
fig = px.scatter(df, x='sepal_width', y='sepal_length',
color='species', hover_data=['petal_length', 'petal_width'])
fig.update_layout(title='Interactive Scatter Plot of Iris Dataset', xaxis_title='Sepal Width', yaxis_title='Sepal Length')
fig.show()
```

Practical Assignment 4: Heatmap with Seaborn
Objective: Generate a heatmap for a correlation matrix.
Steps:
1. Data Preparation:
• Use the Iris dataset available in Seaborn.
• Compute the correlation matrix.
2. Plotting:
• Use Seaborn to create a heatmap.
• Annotate the heatmap with correlation values.
• Add a title.

```python
df = sns.load_dataset('iris')
```

```python
# Compute correlation matrix
corr_matrix = df.select_dtypes(include=np.number).corr() # Only include numerical features for correlation calculation
# Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Iris Dataset')
plt.show()
```

Lab Exercise 1: Customizing Plots
Objective: Customize a plot with additional features.
Steps:
1. Data Preparation:
• Use the same data from Practical Assignment 1.
2. Plotting:
• Create a line plot.
• Add a secondary y-axis with a different scale (e.g., humidity).
• Add a legend for both y-axes.
• Add a title and labels for both y-axes.

```python
import matplotlib.pyplot as plt
# Sample data
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday',
'Saturday', 'Sunday']
temperatures = [22, 24, 21, 23, 25, 26, 27]
humidity = [60, 65, 63, 62, 64, 66, 68]
# Line plot
fig, ax1 = plt.subplots()
ax1.set_xlabel('Day of the Week')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(days, temperatures, label='Temperature',
color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(days, humidity, label='Humidity', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.show()
```

Lab Exercise 2: Exploring Different Plot Types
Objective: Explore different plot types using Seaborn.
Steps:
1. Data Preparation:
• Use the Iris dataset available in Seaborn.
2. Plotting:
• Create a pair plot.
• Create a box plot.
• Create a violin plot.
• Add titles and labels as needed.


```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load data
df = sns.load_dataset('iris')
# Pair plot
sns.pairplot(df, hue='species')
plt.show()
# Box plot
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Box Plot of Sepal Length by Species')
plt.show()
# Violin plot
sns.violinplot(x='species', y='sepal_width', data=df)
plt.title('Violin Plot of Sepal Width by Species')
plt.show()
```
